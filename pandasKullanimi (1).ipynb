```python
import pandas as pd


# Veri setini okuyalÄ±m
file_path = "Grocery_Inventory_and_Sales_Dataset.csv"
df = pd.read_csv(file_path)
df
```

```python
# 1. head() - Veri setinin ilk 5 satÄ±rÄ±nÄ± gÃ¶rÃ¼ntÃ¼leme
print("1. Ä°lk 5 SatÄ±r:")
print(df.head())
```

```python
# 2. info() - Veri setinin yapÄ±sÄ±nÄ± inceleme
print("\n2. Veri Seti Bilgileri:")
print(df.info())
```

```python
# 3. describe() - SayÄ±sal deÄŸiÅŸkenler iÃ§in Ã¶zet istatistikler
print("\n3. SayÄ±sal DeÄŸiÅŸkenlerin Ä°statistikleri:")
print(df.describe())
```

```python
# 4. fillna() - Eksik deÄŸerleri doldurma
df_filled = df.fillna("Bilinmiyor")
print("\n4. Eksik DeÄŸerler Dolduruldu:")
print(df_filled.head())
```

```python
# 5. dropna() - Eksik deÄŸer iÃ§eren satÄ±rlarÄ± silme
df_dropped = df.dropna()
print("\n5. Eksik DeÄŸer Ä°Ã§eren SatÄ±rlar Silindi:")
print(df_dropped.head())
```

```python
# 6. groupby() - Kategori bazlÄ± stok ortalamasÄ±
category_group = df.groupby("Catagory")["Stock_Quantity"].mean()
print("\n6. Kategori BazlÄ± Ortalama Stok MiktarÄ±:")
print(category_group)
```

```python

# Unit_Price sÃ¼tununda boÅŸ veya hatalÄ± deÄŸerleri 'NaN' ile deÄŸiÅŸtirme
df["Unit_Price"] = df["Unit_Price"].replace("[\$,]", "", regex=True)  # Dolar iÅŸaretini kaldÄ±rma
df["Unit_Price"] = df["Unit_Price"].replace(r"^\s*$", "NaN", regex=True)  # BoÅŸ hÃ¼creleri NaN yapma

# NaN olan hÃ¼creleri '0' ya da baÅŸka uygun bir deÄŸerle doldurabilirsiniz
df["Unit_Price"] = pd.to_numeric(df["Unit_Price"], errors='coerce')  # 'NaN' deÄŸerleri sayÄ±sal hale getir

# Gerekirse, NaN olan deÄŸerleri 0 veya baÅŸka bir deÄŸerle doldurabilirsiniz
df["Unit_Price"].fillna(0, inplace=True)

# Stock_Quantity ve Unit_Price'i kullanarak Total_Value hesaplama
df["Total_Value"] = df.apply(lambda row: row["Stock_Quantity"] * row["Unit_Price"], axis=1)

print("\n7. ÃœrÃ¼nlerin Toplam DeÄŸeri HesaplandÄ±:")
print(df.head())

```

```python
# 8. sort_values() - Stok miktarÄ±na gÃ¶re sÄ±ralama
df_sorted = df.sort_values(by="Stock_Quantity", ascending=False)
print("\n8. Stok MiktarÄ±na GÃ¶re SÄ±ralama:")
print(df_sorted.head())
```

```python
# 9. loc[] - Belirli bir koÅŸula gÃ¶re filtreleme
backordered_products = df.loc[df["Status"] == "Backordered"]
print("\n9. SipariÅŸi Bekleyen ÃœrÃ¼nler:")
print(backordered_products.head())
```

```python
# 10. iloc[] - Ä°ndeks bazlÄ± seÃ§im
print("\n10. Ä°lk 5 SatÄ±rÄ±n Ä°ndeks BazlÄ± SeÃ§imi:")
print(df.iloc[0:5])
```
