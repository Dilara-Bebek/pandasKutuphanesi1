```python
import pandas as pd


# Veri setini okuyalım
file_path = "Grocery_Inventory_and_Sales_Dataset.csv"
df = pd.read_csv(file_path)
df
```

```python
# 1. head() - Veri setinin ilk 5 satırını görüntüleme
print("1. İlk 5 Satır:")
print(df.head())
```

```python
# 2. info() - Veri setinin yapısını inceleme
print("\n2. Veri Seti Bilgileri:")
print(df.info())
```

```python
# 3. describe() - Sayısal değişkenler için özet istatistikler
print("\n3. Sayısal Değişkenlerin İstatistikleri:")
print(df.describe())
```

```python
# 4. fillna() - Eksik değerleri doldurma
df_filled = df.fillna("Bilinmiyor")
print("\n4. Eksik Değerler Dolduruldu:")
print(df_filled.head())
```

```python
# 5. dropna() - Eksik değer içeren satırları silme
df_dropped = df.dropna()
print("\n5. Eksik Değer İçeren Satırlar Silindi:")
print(df_dropped.head())
```

```python
# 6. groupby() - Kategori bazlı stok ortalaması
category_group = df.groupby("Catagory")["Stock_Quantity"].mean()
print("\n6. Kategori Bazlı Ortalama Stok Miktarı:")
print(category_group)
```

```python

# Unit_Price sütununda boş veya hatalı değerleri 'NaN' ile değiştirme
df["Unit_Price"] = df["Unit_Price"].replace("[\$,]", "", regex=True)  # Dolar işaretini kaldırma
df["Unit_Price"] = df["Unit_Price"].replace(r"^\s*$", "NaN", regex=True)  # Boş hücreleri NaN yapma

# NaN olan hücreleri '0' ya da başka uygun bir değerle doldurabilirsiniz
df["Unit_Price"] = pd.to_numeric(df["Unit_Price"], errors='coerce')  # 'NaN' değerleri sayısal hale getir

# Gerekirse, NaN olan değerleri 0 veya başka bir değerle doldurabilirsiniz
df["Unit_Price"].fillna(0, inplace=True)

# Stock_Quantity ve Unit_Price'i kullanarak Total_Value hesaplama
df["Total_Value"] = df.apply(lambda row: row["Stock_Quantity"] * row["Unit_Price"], axis=1)

print("\n7. Ürünlerin Toplam Değeri Hesaplandı:")
print(df.head())

```

```python
# 8. sort_values() - Stok miktarına göre sıralama
df_sorted = df.sort_values(by="Stock_Quantity", ascending=False)
print("\n8. Stok Miktarına Göre Sıralama:")
print(df_sorted.head())
```

```python
# 9. loc[] - Belirli bir koşula göre filtreleme
backordered_products = df.loc[df["Status"] == "Backordered"]
print("\n9. Siparişi Bekleyen Ürünler:")
print(backordered_products.head())
```

```python
# 10. iloc[] - İndeks bazlı seçim
print("\n10. İlk 5 Satırın İndeks Bazlı Seçimi:")
print(df.iloc[0:5])
```
